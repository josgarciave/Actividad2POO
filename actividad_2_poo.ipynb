```python
# Clase persona
class Persona:
    def __init__(self, nombre="", apellido="", documento="",
                 año_nacimiento=0):
        # Constructor clase persona
        self.nombre = nombre
        self.apellido = apellido
        self.documento = documento
        self.año_nacimiento = año_nacimiento

    def mostrar_info(self):
        print(f"Nombre = {self.nombre}")
        print(f"Apellido = {self.apellido}")
        print(f"Documento = {self.documento}")
        print(f"Año de nacimiento = {self.año_nacimiento}")
        print("-" * 25)

# Creación de instancias de Persona
if __name__ == "__main__":
    persona1 = Persona("Pedro", "Pérez", "1053121010", 1998)
    persona2 = Persona("Luis", "León", "1053223344", 2001)

    # Mostrar info
    persona1.mostrar_info()
    persona2.mostrar_info()
```

# Ejercicio 2

```python
from enum import Enum

# Enumeración
class TipoPlaneta(Enum):
    GASEOSO = 1
    TERRESTRE = 2
    ENANO = 3

# Clase Planeta
class Planeta:
    def __init__(self, nombre, satelites, masa, volumen, diametro, distancia_sol, tipo_planeta, visible):
        self.nombre = nombre
        self.cantidad_satelites = satelites
        self.masa_kg = masa
        self.volumen_km3 = volumen
        self.diametro_km = diametro
        self.distancia_media_sol_km = distancia_sol
        self.tipo_planeta = tipo_planeta
        self.observable_simple_vista = visible

    def mostrar_info(self):
        print("\n" + "=" * 50)
        print(f"Nombre = {self.nombre}")
        print(f"Satélites = {self.cantidad_satelites}")
        print(f"Masa = {self.masa_kg:.2e} kg")
        print(f"Volumen = {self.volumen_km3:.2e} km³")
        print(f"Diámetro= {self.diametro_km} km")
        print(f"Distancia al Sol = {self.distancia_media_sol_km} millones de km")
        print(f"Tipo = {self.tipo_planeta.name if self.tipo_planeta else 'No especificado'}")
        print(f"Visible a simple vista = {True if self.observable_simple_vista else False}")

# Creación de planetas
def main():
    planetas = [
        Planeta(
            nombre="Tierra",
            satelites=1,
            masa=5.9736E24,
            volumen=1.08321E12,
            diametro=12742,
            distancia_sol=150000000,
            tipo_planeta=TipoPlaneta.TERRESTRE,
            visible=True
        ),
        Planeta(
            nombre="Júpiter",
            satelites=79,
            masa=1.899E27,
            volumen=1.4313E15,
            diametro=139820,
            distancia_sol=750000000,
            tipo_planeta=TipoPlaneta.GASEOSO,
            visible=True
        )
    ]

    # Mostrar información de los planetas
    for planeta in planetas:
        planeta.mostrar_info()

if __name__ == "__main__":
    main()
```

# Ejercicio 3

```python
from enum import Enum

# Enumeración
class TipoCombustible(Enum):
    GASOLINA = "GASOLINA"
    BIOETANOL = "BIOETANOL"
    DIESEL = "DIESEL"
    BIODIESEL = "BIODIESEL"
    GAS_NATURAL = "GAS_NATURAL"

class TipoAutomovil(Enum):
    CARRO_CIUDAD = "CARRO_CIUDAD"
    SUBCOMPACTO = "SUBCOMPACTO"
    COMPACTO = "COMPACTO"
    FAMILIAR = "FAMILIAR"
    EJECUTIVO = "EJECUTIVO"
    SUV = "SUV"

class Color(Enum):
    BLANCO = "BLANCO"
    NEGRO = "NEGRO"
    ROJO = "ROJO"
    NARANJA = "NARANJA"
    AMARILLO = "AMARILLO"
    VERDE = "VERDE"
    AZUL = "AZUL"
    VIOLETA = "VIOLETA"

# Clase Automovil
class Automovil:
    def __init__(self, marca: str, modelo: int, motor: int,
                 tipo_combustible: TipoCombustible, tipo_automovil: TipoAutomovil,
                 numero_puertas: int, cantidad_asientos: int, velocidad_maxima: int,
                 color: Color, velocidad_actual: int = 0):

        # Setters para iniciar los atributos
        self.set_marca(marca)
        self.set_modelo(modelo)
        self.set_motor(motor)
        self.set_tipo_combustible(tipo_combustible)
        self.set_tipo_automovil(tipo_automovil)
        self.set_numero_puertas(numero_puertas)
        self.set_cantidad_asientos(cantidad_asientos)
        self.set_velocidad_maxima(velocidad_maxima)
        self.set_color(color)
        self.set_velocidad_actual(velocidad_actual)

    # Getters y Setters
    def get_marca(self) -> str:
        return self._marca

    def set_marca(self, marca: str):
        self._marca = marca

    def get_modelo(self) -> int:
        return self._modelo

    def set_modelo(self, modelo: int):
        self._modelo = modelo

    def get_motor(self) -> int:
        return self._motor

    def set_motor(self, motor: int):
        if motor > 0:
            self._motor = motor
        else:
            print("Error: El motor debe ser positivo.")

    def get_tipo_combustible(self) -> TipoCombustible:
        return self._tipo_combustible

    def set_tipo_combustible(self, tipo_combustible: TipoCombustible):
        if isinstance(tipo_combustible, TipoCombustible):
            self._tipo_combustible = tipo_combustible
        else:
            print("Error: Tipo de combustible no válido.")

    def get_tipo_automovil(self) -> TipoAutomovil:
        return self._tipo_automovil

    def set_tipo_automovil(self, tipo_automovil: TipoAutomovil):
        if isinstance(tipo_automovil, TipoAutomovil):
            self._tipo_automovil = tipo_automovil
        else:
            print("Error: Tipo de automóvil no válido.")

    def get_numero_puertas(self) -> int:
        return self._numero_puertas

    def set_numero_puertas(self, numero_puertas: int):
        if numero_puertas > 0:
            self._numero_puertas = numero_puertas
        else:
            print("Error: El número de puertas debe ser positivo.")

    def get_cantidad_asientos(self) -> int:
        return self._cantidad_asientos

    def set_cantidad_asientos(self, cantidad_asientos: int):
        if cantidad_asientos > 0:
            self._cantidad_asientos = cantidad_asientos
        else:
            print("Error: La cantidad de asientos debe ser positiva.")

    def get_velocidad_maxima(self) -> int:
        return self._velocidad_maxima

    def set_velocidad_maxima(self, velocidad_maxima: int):
        if velocidad_maxima > 0:
            self._velocidad_maxima = velocidad_maxima
        else:
            print("Error: La velocidad máxima debe ser positiva.")

    def get_color(self) -> Color:
        return self._color

    def set_color(self, color: Color):
        if isinstance(color, Color):
            self._color = color
        else:
            print("Error: Color no válido.")

    def get_velocidad_actual(self) -> int:
        return self._velocidad_actual

    def set_velocidad_actual(self, velocidad_actual: int):
        if 0 <= velocidad_actual <= self._velocidad_maxima:
            self._velocidad_actual = velocidad_actual
        else:
            print(f"Error: La velocidad debe estar entre 0 y {self._velocidad_maxima} km/h.")

    # Métodos conducción
    def acelerar(self, incremento):
      nueva_velocidad = self._velocidad_actual + incremento
      if nueva_velocidad > self._velocidad_maxima:
        self._velocidad_actual = nueva_velocidad
        print(f"Velocidad actual: {self._velocidad_actual} km/h")

    def desacelerar(self, decremento: int):
        nueva_velocidad = self._velocidad_actual - decremento
        if nueva_velocidad < 0:
            print("No se puede decrementar a una velocidad negativa.")
            self.set_velocidad_actual(0)

    def frenar(self):
        self.set_velocidad_actual(0)
        print(f"Velocidad actual = {self._velocidad_actual}")

    #  Método para mostrar información
    def mostrar_informacion(self):
        print(f"Marca = {self._marca}")
        print(f"Modelo = {self._modelo}")
        print(f"Motor = {self._motor}")
        print(f"Tipo de combustible = {self._tipo_combustible.value}")
        print(f"Tipo de automóvil = {self._tipo_automovil.value}")
        print(f"Número de puertas = {self._numero_puertas}")
        print(f"Cantidad de asientos = {self._cantidad_asientos}")
        print(f"Velocidad máxima = {self._velocidad_maxima} km/h")
        print(f"Color = {self._color.value}")
        print(f"Velocidad actuala = {self._velocidad_actual}")
        print(f"")

#  instancia y ejecución
auto1 = Automovil(
    marca="Ford",
    modelo=2018,
    motor=3,
    tipo_combustible=TipoCombustible.DIESEL,
    tipo_automovil=TipoAutomovil.EJECUTIVO,
    numero_puertas=5,
    cantidad_asientos=6,
    velocidad_maxima=250,
    color=Color.NEGRO,

)

# Mostrar información y probar métodos
auto1.mostrar_informacion()
```

# Ejercicio 4

```python
import math

class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * math.pow(self.radio, 2)

    def perimetro(self):
        return 2 * math.pi * self.radio

class Rectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def perimetro(self):
        return 2 * (self.base + self.altura)

class Cuadrado:
    def __init__(self, lado):
        self.lado = lado

    def area(self):
        return math.pow(self.lado,2)

    def perimetro(self):
        return 4 * self.lado

class TrianguloRectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return (self.base * self.altura) / 2

    def perimetro(self):
        return self.base + self.altura + self.hipotenusa()

    def hipotenusa(self):
        return math.sqrt(self.base ** 2 + self.altura ** 2)

    def tipo_triangulo(self):
        base = round(self.base, 5)
        altura = round(self.altura, 5)
        hipotenusa = round(self.hipotenusa(), 5)

        if base == altura == hipotenusa:
            return "Equilátero"
        elif base == altura or base == hipotenusa or altura == hipotenusa:
            return "Isósceles"
        else:
            return "Escaleno"

def main():
    circulo = Circulo(2)
    print("Círculo:")
    print(f"Área: {circulo.area():.2f} cm²")
    print(f"Perímetro: {circulo.perimetro():.2f} cm\n")

    rectangulo = Rectangulo(1, 2)
    print("Rectángulo:")
    print(f"Área: {rectangulo.area()} cm²")
    print(f"Perímetro: {rectangulo.perimetro()} cm\n")

    cuadrado = Cuadrado(3)
    print("Cuadrado:")
    print(f"Área: {cuadrado.area()} cm²")
    print(f"Perímetro: {cuadrado.perimetro()} cm\n")

    triangulo = TrianguloRectangulo(3, 5)
    print("Triángulo Rectángulo:")
    print(f"Área: {triangulo.area()} cm²")
    print(f"Perímetro: {triangulo.perimetro():.2f} cm")
    print(f"Hipotenusa: {triangulo.hipotenusa():.2f} cm")
    print(f"Tipo de triángulo: {triangulo.tipo_triangulo()}\n")

if __name__ == "__main__":
    main()
```

# Ejercicio 5

```python
from enum import Enum

class TipoCuenta(Enum):
    AHORROS = "Ahorros"
    CORRIENTE = "Corriente"

class CuentaBancaria:
    def __init__(self, nombres_titular, apellidos_titular, numero_cuenta, tipo_cuenta):
        self.nombres_titular = nombres_titular
        self.apellidos_titular = apellidos_titular
        self.numero_cuenta = numero_cuenta
        self.tipo_cuenta = tipo_cuenta
        self.saldo = 0.0

    def imprimir(self):
        print(f"Nombres del titular: {self.nombres_titular}")
        print(f"Apellidos del titular: {self.apellidos_titular}")
        print(f"Número de cuenta: {self.numero_cuenta}")
        print(f"Tipo de cuenta: {self.tipo_cuenta.value}")
        print(f"Saldo: ${self.saldo:.2f}")

    def consultar_saldo(self):
        print(f"El saldo actual es: ${self.saldo:.2f}")

    def consignar(self, valor):
        if valor > 0:
            self.saldo += valor
            print(f"Se ha consignado ${valor:.2f}. El nuevo saldo es ${self.saldo:.2f}.")
            return True
        else:
            print("El valor a consignar debe ser mayor que cero.")
            return False

    def retirar(self, valor):
        if valor > 0 and valor <= self.saldo:
            self.saldo -= valor
            print(f"Se ha retirado ${valor:.2f}. El nuevo saldo es ${self.saldo:.2f}.")
            return True
        else:
            print("El valor a retirar debe ser mayor que cero y no superar el saldo actual.")
            return False

# Creación de cuenta bancaria
if __name__ == "__main__":
    cuenta = CuentaBancaria("Pedro", "Pérez", 123456789, TipoCuenta.AHORROS)
    cuenta.imprimir()
    cuenta.consignar(200000)
    cuenta.consignar(300000)
    cuenta.retirar(400000)
```
